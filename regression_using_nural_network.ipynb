# Regression model using Deep learning


```python
import tensorflow as tf
import keras
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import layers
# TO display all the columns
pd.pandas.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('x_train_final.csv')
df2=pd.read_csv('y_train_final.csv')
```

```python
df.head()
x=df
```

```python
df2=df2.iloc[:,1:]
df2.head()
y=df2
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
y_train.head()
```

## https://www.youtube.com/watch?v=-vHQub0NXI4

```python
def build_model():
    model=keras.Sequential([
        layers.Dense(60,activation='relu',input_shape=[x_train.iloc[1].size]),
        layers.Dense(50,activation='relu'),
        layers.Dense(1)
        
        ])
    optimizer=tf.keras.optimizers.RMSprop(0.001)
    model.compile(loss='mse',
             optimizer=optimizer,
             metrics=['mae','mse'])

    return model
```

```python
model=build_model()
```

```python
model.summary()
```

## Now lets try

```python
history=model.fit(x_train,y_train,
                 epochs=1000,validation_split=0.2,verbose=1)
```

```python
hist=pd.DataFrame(history.history)
```

```python
hist.tail()
```

```python
plt.plot(hist['val_mse'].iloc[200:])
plt.plot(hist['mse'].iloc[200:])
```
