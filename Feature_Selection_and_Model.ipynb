# Feature Selection

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
# to visualize all the columns in dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
dataset=pd.read_csv('x_train_final.csv')
x_train=dataset
```

```python
dataset.head()
```

```python
y_train=pd.read_csv('y_train_final.csv')
y_train=y_train.iloc[:,1:]
y_train.head()
```

```python
# now we will apply feature selection
from sklearn.linear_model import Lasso
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.feature_selection import SelectFromModel
```

```python
x_t,x_test,y_t,y_test=train_test_split(x_train,y_train,test_size=0.33,random_state=0)
```

```python
pipeline=Pipeline([('model',Lasso())
                  ])
```

```python
search=GridSearchCV(pipeline,{'model__alpha':np.arange(0.1,10,0.1)},cv=5,scoring="neg_mean_squared_error",verbose=3)
```

```python
search.fit(x_t,y_t)
```

```python
search.best_params_
```

## Now we have found the best alpha for lasso regression

### one thing to note is that  

#### 1. The bigger the alpha value , the lesser number of features will get selected

# Now we will run Feature selection_model

```python
feature_selection_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_selection_model.fit(x_t,y_t)
```

## We can actually tweak with alpha and see , how the feature selection is affected

```python
feature_selection_model.get_support()
```

```python
#Now to select the features
selected_features=x_t.columns[feature_selection_model.get_support()]
selected_features
```

```python
print(" The total features were " , len(x_t.columns))
print('selected_features were', len(selected_features))
```
